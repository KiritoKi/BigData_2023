```python
from database.session import engine
import pandas as pd
import matplotlib.pyplot as plt

query = "SELECT * FROM grupo_de_pesquisa;"

df = pd.read_sql(query, engine)

grupo_por_area = df.groupby('grande_area_predominante').size()

grupo_por_area.plot(kind='bar')
plt.title('Contagem por Grande rea Predominante')
plt.xlabel('Grande rea Predominante')
plt.ylabel('Contagem')
plt.show()
```
