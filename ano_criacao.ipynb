```python
from database.session import engine
import pandas as pd
import matplotlib.pyplot as plt

query = "SELECT * FROM grupo_de_pesquisa;"

df = pd.read_sql(query, engine)

grupo_por_area = df.groupby('ano_de_criacao').size()


plt.figure(figsize=(12, 6)) 

grupo_por_area.plot(kind='bar')
plt.title('Contagem por Ano de Criação')
plt.xlabel('Ano de Criação')
plt.ylabel('Contagem')
plt.show()
```
