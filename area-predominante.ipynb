```python
from database.session import engine
import pandas as pd
import matplotlib.pyplot as plt

query = "SELECT * FROM grupo_de_pesquisa;"

df = pd.read_sql(query, engine)

grupo_por_area = df.groupby('area_predominante').size()


plt.figure(figsize=(15, 6)) 

grupo_por_area.plot(kind='bar')
plt.title('Contagem por Área Predominante')
plt.xlabel('Área Predominante')
plt.ylabel('Contagem')
plt.show()
```
